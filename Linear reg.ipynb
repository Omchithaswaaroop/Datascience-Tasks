```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("Housing.csv")
data
```

```python
data.columns
```

```python
price=data['price']
price
```

```python
area=data['area']
area
```

```python
x=np.array(price).reshape(-1,1)
y=np.array(area).reshape(-1,1)
```

```python
x
```

```python
y
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)## Training The Model
```

```python
new_area=int(input("Enter a Area to Predict the price:"))
new_data=np.array(new_area).reshape(-1,1)
y_answer=model.predict(new_data)
print(F"Price of {new_area} area is:",model.predict(new_data))
```

```python
plt.plot(x,y)
plt.scatter(new_data,y_answer)
plt.xlabel("Price")
plt.ylabel('Area')
plt.show()
```
