```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
house_value=data['median_house_value']
house_value
```

```python
from scipy.stats import skew
skew_value=house_value.skew()
```

```python
print("skew Value:",skew_value)

if skew_value>0:
    print("positive skew")
elif skew_value<0:
    print("negative skew")
else:
    print("normal skew")
```

```python
plt.hist(house_value, color='pink', bins=10)
plt.title("House Value Distribution")
plt.xlabel("House value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
