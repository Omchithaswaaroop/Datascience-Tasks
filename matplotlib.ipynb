```python
data3 = pd.read_csv("netflix_titles.csv")
data3
```

```python
data5 = pd.read_csv("daily_weather.csv")
data5
```

```python
temp = data5["air_temp_9am"]
humidity = data5["relative_humidity_9am"]
windspeed = data5["avg_wind_speed_9am"]

plt.boxplot([temp, humidity, windspeed],
            labels=["Temperature", "Humidity", "Wind Speed"])

plt.title("Weather Analysis")

plt.show()
```
