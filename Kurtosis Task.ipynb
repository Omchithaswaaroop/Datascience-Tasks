```python
Monthly_income=HR_data['MonthlyIncome']
Monthly_income
```

```python
Kurtosis_value=Monthly_income.kurt()
```

```python
print("Kurtosis Value=",Kurtosis_value)

if Kurtosis_value > 3:
    print("LeptoKurtic Distribution -> High Outliers -> Need to investigate the data")
elif Kurtosis_value < 3:
     print("PlatyKurtic Distribution -> Low Outliers -> Need to investigate the data")
elif Kurtosis_value == 3:
     print("MesoKurtic Distribution -> No Outliers -> Need to investigate the data")
```

```python
plt.boxplot(Monthly_income)
plt.title("Monthly Income")
plt.show()
```
