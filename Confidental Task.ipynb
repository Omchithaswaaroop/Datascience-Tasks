```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('Employee_Salary_Dataset.csv')
data
```

```python
salary=data["Salary"]

population_mean = np.mean(salary)

print("Population Mean Salary =", population_mean)
```

```python
salary_data=data['Salary'].head(15)
sample_data=np.array(salary_data)
sample_data
```

```python
from scipy import stats
sample_mean = np.mean(sample_data)

## T- distribution because data is less than 30.
ci_cal = stats.t.interval( ## T distribution
    confidence=0.95, ## 95% confident
    df=len(sample_data)-1, ## Degree of freedom
    loc=sample_mean, 
    scale=stats.sem(sample_data)
)
print("Sample Mean =", sample_mean)
print("95% Confidence Interval =", ci_cal)
```

```python
from scipy.stats import norm
salary_data_2=data['Salary']
sample_data_2=np.array(salary_data_2)
sample_data_2
```

```python
sample_mean = np.mean(sample_data_2)

sample_std = np.std(sample_data_2, ddof=1)

n = len(sample_data_2)

z = 1.96

margin_error = z * (sample_std / np.sqrt(n))

lower_ci = sample_mean - margin_error
upper_ci = sample_mean + margin_error

print("Sample Size =", n)
print("Sample Mean =", round(sample_mean,2))
print("Sample Standard Deviation =", round(sample_std,2))
print("Margin of Error =", round(margin_error,2))

print("\n95% Z-Confidence Interval")
print("Lower Bound =", round(lower_ci,2))
print("Upper Bound =", round(upper_ci,2))
     
```

```python
from scipy.stats import norm
# Given values
mu = 100  # mean
sigma = 15  # standard deviation
X = 90-110  # value to check
prob_greater = 1 - norm.cdf(X, mu, sigma)
print("Probability:", round(prob_greater * 100, 2), "%")
```

```python
## Normal Distribution
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu = 100
sigma = 15
X = np.random.normal(mu, sigma, 10000)

plt.hist(X, bins=30, density=True, alpha=0.6)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, linewidth=2)

plt.title("Normal Distribution Histogram\nBell Shaped Curve")
plt.xlabel("IQ")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
mean_time = 10  
t = 3  

lambda_value = 1 / mean_time
prob_within = expon.cdf(t, scale=1/lambda_value)

print("Probability customer arrives within 3 minutes:",
      round(prob_within * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_time = 10  # average customer arrival time

data = np.random.exponential(mean_time, 10000)

plt.hist(data, bins=30, density=True)  

plt.title("Exponential Distribution - Histogram")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
