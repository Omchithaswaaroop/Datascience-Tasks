```python
price_data=data['price']
price_data
```

```python
price_data_100=price_data[:101]
price_data_100
```

```python
kurtosis_value=price_data_100.kurt()
```

```python
print("Kurtosis Value:", kurtosis_value)

if kurtosis_value > 3:
    print("LeptoKurtic Distribution")
elif kurtosis_value < 3:
     print("PlatyKurtic Distribution")
elif kurtosis_value == 3:
     print("MesoKurtic Distribution")
    
```

```python
plt.hist(price_data_100, bins=20)
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(price_data_100)
plt.title("House Price Distribution")
plt.show()
```
